### 001.001 Simple Generator

1. Make `countdown` a generator that sits and waits until you call for the next number


```python
# def countdown(n):
#     print("Counting down from", n)
#     print()
#     print("Done counting down")


# # Example use
# for i in countdown(10):
#     print(i, end=" ")

# solution


def countdown(n):
    print("Counting down from", n)
    while n > 0:
        yield n
        n -= 1
    print()
    print("Done counting down")


# Example use
for i in countdown(10):
    print(i, end=" ")
```
