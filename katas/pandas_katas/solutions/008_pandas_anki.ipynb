### 008.001 Extract 1st sentence

1. Read file_last_step and print the head
1. Add a column "s3" after "s2" (value=""), without knowing where s2 is
1. Create a function `extract_first_sentence` that applies to each row, and
   1. Takes the text from  `column_name` and if there..
   1. Removes any comma which follows `punctuation` (try to use lookbehind)
   1. Use nltk to split into sentences (it's german), and make sure there are at least 3 of them (empty if not there)
   1. If s1 is empty put that sentence there, if not put it in s2, and so on - s1, s2, s3 should all be filled if possible
   1. Run it on all rows (REMEMBER THE USUAL ERROR)
   1. show its head
1. Removes `column_name` and
   1. add empty column "s1e" ... "s3e" after each of "s1"..."s3", without knowing where they are
   1. show the first 20 rows
1. Writes to `save_to`
   1. without index column
   1. Print where you saved it

```python

import re
import nltk
nltk.download('punkt')

file_last_step = STEP_1

column_name = 'Sample sentence Typing Corrects'
punctuation = r"[,.!?]"
save_to = STEP_2

1
df = pd.read_csv(file_last_step, sep="\t").fillna("")
df.head()
# solution

2
df.insert(df.columns.get_loc("s2") + 1, "s3", "")
df.head()

3
def extract_first_sentence(row):
    text = row[column_name]
    if not text:
        return row
    text = re.sub(rf"(?<={punctuation}),", "", text)
    sentences = nltk.sent_tokenize(text, language="german") + 3 * [""]
    for s in ["s1", "s2", "s3"]:
        if not row[s]:
            row[s] = sentences.pop(0)
    return row

df = df.apply(extract_first_sentence, axis="columns")
df.head()

4
df.drop(columns=[column_name], inplace=True)
for s in ["s1", "s2", "s3"]:
    df.insert(df.columns.get_loc(s) + 1, s + "e", "")
df[:20]

5
df.to_csv(save_to, sep="\t", index=False)
print(f"Saved to {save_to}")


```

### 006.003 Rearrange columns

1. Read `file_last_step` into `merged_df` with 
   1. 0th col as index
   1. Use "" as empty cell value
   1. Print the head
1. Create a function `move_across` which takes a row as input and
   1. if `s1` is empty, copies `s1` and `s2` and `s3` from goethe
   1. if `s2` is empty, copies `s1` and `s2` from goethe to `s2` and `s3` respecfively
   1. if `s3` is empty, copies `s1` from goethe to it 
   1. Uses values from goethe as default if those from original are empty, for columns `default_columns`
   1. Apply to all rows
1. Writes to `save_to`
   1. This time with index column
   1. Print where you saved it

```python
file_last_step = STEP_3

default_columns = ["artikel_d", "full_d", "base_e"]

save_to = STEP_4

1
merged_df = pd.read_csv(file_last_step, sep="\t", index_col=0).fillna("")
merged_df.head()
# solution

2
default_columns_from = { s: f"{s}{suffix}" for s in default_columns}
cols_to = ["s1", "s2", "s3"]

def move_across(row):
    cols_from = [f"{s}{suffix}" for s in cols_to]
    for s in cols_to:
        if not row[s]:
            key = cols_from.pop(0)
            row[s] = row[key]
    for s in default_columns:
        if not row[s]:
            row[s] = row[default_columns_from[s]]
    return row

merged_df = merged_df.apply(move_across, axis="columns")
merged_df.head()

3
merged_df.to_csv(save_to, sep="\t")
print(f"Saved to {save_to}")

```
