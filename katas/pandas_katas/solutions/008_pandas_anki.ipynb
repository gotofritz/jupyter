### 006.004 Merge DF

1. Read `file_last_step` into `df_4000_words_4` with 
   1. 2nd col as index
   1. Use "" as empty cell value
   1. Print the head
1. Read `file_to_merge` into `df_goethe`
   1. 2nd col as index
   1. Use "" as empty cell value
   1. Note that it doesn't have a header row
   1. Copy the column name from `df_4000_words_4`
   1. Print the head
1. Merge `df_goethe` into the other and save as `merged_df`
   1. both have index
   1. merge should have all the rows from both dfs
   1. Use `suffix` as the suffix for `df_goethe`, and none for the other
   1. Use "" as empty cell
   1. Show the head
1. Create a function `move_across` which takes a row as input and
   1. if `s1` is empty, copies `s1` and `s2` and `s3` from goethe
   1. if `s2` is empty, copies `s1` and `s2` from goethe to `s2` and `s3` respecfively
   1. if `s3` is empty, copies `s1` from goethe to it 
   1. Uses values from goethe as default if those from original are empty, for columns `default_columns`
   1. Apply to all rows
1. Remove all columns with `suffix` in the name
1. Prepare to write to csv, by removing the index so `base_d` is back at position 2
   1. Remove the index
   1. Take it out and reinsert it in the right place
   1. Print the first 20 rows

```python
file_last_step = "006_4000_words_step_2.tsv"
df_4000_words_4 = None

file_to_merge = "006_Goethe_words.tsv"
df_goethe = None

suffix = '_goethe'
default_columns = ["artikel_d", "full_d", "base_e"]
save_to = "006_merged.tsv"

# solution

1
df_4000_words_4 = pd.read_csv(file_last_step, sep="\t", index_col=2).fillna("")
df_4000_words_4.head()

2
df_goethe = pd.read_csv(file_to_merge, sep="\t", index_col=2, header=None, names=df_4000_words_4.columns).fillna("")
df_goethe.head()

3
merged_df = df_4000_words_4.merge(df_goethe, left_index=True, right_index=True, how="outer", suffixes=["", suffix]).fillna("")
merged_df.head()

4
def move_across(row):
    if row["s1"] == "":
        row["s1"] = row[f"s1{suffix}"]
        row["s2"] = row[f"s2{suffix}"]
        row["s3"] = row[f"s3{suffix}"]
    elif row["s2"] == "":
        row["s2"] = row[f"s1{suffix}"]
        row["s3"] = row[f"s2{suffix}"]
    elif row["s3"] == "":
        row["s3"] = row[f"s1{suffix}"]
    for col in default_columns:
        if row[col] == "":
            row[col] = row[f"{col}{suffix}"]
    return row
merged_df = merged_df.apply(move_across, axis=1)
merged_df.head()

5
merged_df.drop(merged_df.filter(like=suffix).columns, axis=1, inplace=True)
merged_df.head()

6
merged_df = merged_df.reset_index()
new_col = merged_df.pop('index')
merged_df.insert(2, 'base_d', new_col)
merged_df[:20]

7
merged_df.to_csv(save_to, sep="\t", index=False, header=True)
```

### 006.005 Add parts of speech

1. Read `file_last_step` into `merged_df` with 
   1. "" as empty cell value
   1. Print the head
1. Create a set called `stop_words` with english stop words
1. Create a function `find_pos` which takes a row as an input, and 
   1. if cell `pos` is full, just return row
   1. if not, take `row["base_e"]` as base and run two regexes on it:
      1. remove all parenthesis and their content
      1. remove anything that is not a letter or space
   1. Create a list of tokens, as long as they are not in stopwords
   1. get 1st pos tag for them and return
   1. apply to `merged_df` and prints 20 rows

```python
import nltk
import re
from nltk.corpus import stopwords

nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

file_last_step = "006_merged.tsv"
merged_df = None

pos = "pos"

stop_words = None

1
merged_df = pd.read_csv(file_last_step, sep="\t").fillna("")
merged_df.head()
# solution

2
stop_words = set(stopwords.words('english'))

3
def find_pos(row):
    if row[pos]:
        return row
    raw = re.sub(r" *\(.+\)", "", row["base_e"])
    raw = re.sub(r"[^a-z ]", "", raw, re.I)
    
    tokens = [token for token in nltk.word_tokenize(raw) if token not in stop_words]
    pos_tags = nltk.pos_tag(tokens)
    
    if pos_tags:
        row[pos] = pos_tags[0][1]
    return row
    
merged_df = merged_df.apply(find_pos, axis=1)
merged_df[:20]

    
```
