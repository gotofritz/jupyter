```python
import pandas as pd
import perfplot

from codetiming import Timer

from IPython.core.interactiveshell import InteractiveShell

pd.set_option('display.max_rows', None)
InteractiveShell.ast_node_interactivity = "all"

STEP_1 = "006_4000_words.tsv"
STEP_2 = "006_4000_words_end.tsv"
```

### 006.001 Create the summary above

1. Read file_4000_words and print the head, bearing in mind that
   1. it's a tsv file
   1. the file has no header row; the column names are in cols_4000_words
   1. empty cells should be an empty string
1. Print the first 10 rows of the table in markdown format (github-flavoured tables)
   1. NOTE: you actually have a print statement
   1. Create a function that truncates it to 17 characters and adds 3 dots, and use the function that applies it to every cell
   1. You should hide the index column when you generate the markdown


```python
file_4000_words = STEP_1
cols_4000_words = ["German", "English", "Plural and inflected forms", "Sample sentence", "Sample sentence Typing Corrects", "Part of Speech", "Audio"]
# solution

1
df = pd.read_csv(file_4000_words, sep="\t", header=None, names=cols_4000_words).fillna("")
df.head()

2
max_len = 20
ellipse = '...'
def trunc(cell):
    if isinstance(cell, str) and len(cell) > (max_len - len(ellipse)):
        return cell[:17] + ellipse
    return cell

print(df[:10].applymap(trunc).to_markdown(index=False, tablefmt="github"))
```

### 006.002 Rearrange columns

1. Read file_4000_words and print the head
1. Create a new df, `name_article_df`
   1. It will end up having two columns: one with the article, if it exists, the other with the noun without the article, or with the complete entry otherwise
   1. take the "German" column as starting point and use a regular expression to extract the two columns
       1. Use named group `artikel_d` for the article followed by a space, and `base_d` for the rest of the string, without article if any
   1. Make sure there are always two column - but you don't need an extra step with apply for that
   1. Make sure NaN are empty strings
   1. print head
1. Insert `base_d` as 2nd, and `artikel_d` as 4th in `df`
1. Use insert to..
   1. rename `col_0` "Plural and inflected forms" to `full_d` and move it to 1st
   1. rename "English" to `base_e` and move it to 2nd
   1. Remove both columns and "German"
   1. rename "Part of Speech" to "pos" and "Audio" to "audio"
1. Create 2 new columns, "s1" and "s2", directly in `ds` by splitting `Sample sentence` on `2) `
   1. Make sure "None" columns are actually empty strings
   1. Make sure only the 1st 2) is matched, if there are more
1. Remove the "1) " prefix from those columns which have it
   1. remove any trailing spaces
   1. Finally, remove the `remove_col` column
1. Create a boolean mask `mask` for those cells in the "pos" column, whose string values contains a bracket `)`
   1. Set the value to ''
   1. Show the first 20 rows
1. Save as tsv to `save_to`, without index

```python

pd.reset_option('display.max_colwidth')

file_4000_words = STEP_1
cols_4000_words = ["German", "English", "Plural and inflected forms", "Sample sentence", "Sample sentence Typing Corrects", "Part of Speech", "Audio"]
col_0 = "Plural and inflected forms"
name_article_df = None
remove_col = "Sample sentence"
save_to = STEP_2

1
df = pd.read_csv(file_4000_words, sep="\t", header=None, names=cols_4000_words).fillna("")
df.head()

# solution

2
name_article_df = df["German"].str.extract(r"^(?P<artikel_d>der\s+|die\s+|das\s+)?(?P<base_d>.+)$", expand=True).fillna("")
name_article_df.head()

3
col = "base_d"
df.insert(2, col, name_article_df[col])
col = "artikel_d"
df.insert(4, col, name_article_df[col])
df.head()

4
col_to = "full_d"
col_from = col_0
df.insert(0, col_to, df[col_from])
df.head()

col_to = "base_e"
col_from = "English"
df.insert(2, col_to, df[col_from])
df.drop(columns=["English", col_0, "German"], inplace=True)
df.rename(columns={"Part of Speech": "pos", "Audio": "audio"}, inplace=True)
df.head()

5
df[["s1", "s2"]] = df["Sample sentence"].str.split(r" 2\)", n=1, expand=True).fillna("")

6
df['s1'] = df['s1'].str.lstrip('1) ').str.strip()
df['s2'] = df['s2'].str.strip()
df.drop(columns=[remove_col], inplace=True)
df.head()

7
mask = df["pos"].str.contains(r'\)')
df.loc[mask, "pos"] = ''
df[:20]

8
df.to_csv(save_to, sep="\t", index=False)


```
